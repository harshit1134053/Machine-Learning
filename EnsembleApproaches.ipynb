```python
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer

def encode_and_impute_columns(df):
    df_encoded = df.copy()
    le = LabelEncoder()
    imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

    # Loop through each column to encode or impute
    for column in df_encoded.columns:
        if df_encoded[column].dtype == 'object':
            # Fill NaN with the string 'Missing' before label encoding
            df_encoded[column] = df_encoded[column].fillna('Missing')
            # Transform the non-numeric columns to numeric labels
            df_encoded[column] = le.fit_transform(df_encoded[column])
        else:

            df_encoded[column] = imputer.fit_transform(df_encoded[[column]])

    return df_encoded
```

```python
input_data = data.iloc[:, :-1]
target_labels = data.iloc[:, -1]
```

```python
training_data, testing_data, training_labels, testing_labels = train_test_split(
    input_data, target_labels, test_size = 0.2, random_state = 1
)
```

```python
len(training_data), len(testing_data)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report
```

```python
random_forest = RandomForestClassifier(n_estimators=100,oob_score=True,max_features=5)

random_forest.fit(training_data, training_labels)

Y_pred = random_forest.predict(testing_data)

random_forest.score(training_data, training_labels)
```

```python
print(classification_report(testing_labels, Y_pred))
```

```python
grad_boost = GradientBoostingClassifier(n_estimators=1000)
grad_boost.fit(training_data, training_labels)
Y_pred = grad_boost.predict(testing_data)
grad_boost.score(training_data, training_labels)
```
