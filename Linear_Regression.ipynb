```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv('/content/data.csv')


x = df['6.1101']
y = df['17.592']
def linear_regression(x, y):

    n = len(x)
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    #here we try to create values for equation y=mx+b
    numerator = np.sum((x - x_mean) * (y - y_mean))
    denominator = np.sum((x - x_mean)**2)
    m = numerator / denominator
    b = y_mean - m * x_mean

    return m, b
def plot_regression_line(x, y, m, b):

    plt.scatter(x, y, color='blue', label='Data Points')


    regression_line = m * x + b
    plt.plot(x, regression_line, color='red', label='straight Line')


    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()

slope, intercept = linear_regression(x, y)
print(f"Equation of the regression line: y = {slope:.2f}x + {intercept:.2f}")

plot_regression_line(x, y, slope ,intercept)
```
